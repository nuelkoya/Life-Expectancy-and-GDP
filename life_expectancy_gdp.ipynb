# Life expectancy and GDP

all_data.csv contains data from 6 countries namely: Chile, China, Germany, Mexico, USA and Zimbabwe. With columns showing how GDP and Life expectancy as changed from 2000 to 2015.

Here are some questions that will be answered using the data:
- Has life expectancy increased over time in the six nations?
- Has GDP increased over time in the six nations?
- Is there a correlation between GDP and life expectancy of a country?
- what is the average life expectancy in each country
- What is the distribution of that life expectancy?
- Size of each nations GDP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv('all_data.csv')
columns = data.columns.values
data_countries = data.Country.unique()
data.head()
```

# Has life expectancy increased over time in the six nations?

In the cell below, we can see that over time(2000 -2015) all nations in the data has experienced an increase in life expextancy of it's people.

```python
for country in data_countries:
    plt.plot("Year", "Life expectancy at birth (years)", data = data[data['Country'] == country])

plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])
plt.title('Change in Life Expectancy between Countries')
plt.xlabel('Year')
plt.ylabel('Life Expectancy(years)')
plt.show()
```

# Has GDP increased over time in the six nations?
We can see that over time(2000 -2015) all nations in the data has experienced an increase in GDP. China saw the most significant increase in GDP by 813.42 percent. Chile had a GDP growth overtime of 211.48 percent. Zimbawbwe had a GDP growth of 143.72 percent. While USA, Germany and Mexico had a GDP growth of 75.73, 73.11, and 68.55 percent respectively.

```python
for country in data_countries:
    plt.plot("Year", "GDP", data = data[data['Country'] == country])
    gdp_list = list(data.GDP[data['Country'] == country])
    gdp_growth_rate = round(((gdp_list[-1] - gdp_list[0]) / gdp_list[0]) * 100, 2)
    print('From 2000 to 2015, {country} had an overall GDP growth of {GDP} percent'.format(country=country, GDP=gdp_growth_rate))
plt.title('GDP Growth')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()

    
    
```

# Is there a correlation between GDP and life expectancy of a country?
There is a very strong linear relationship between GDP and the life expectancy in all 6 countries.


```python
for country in data_countries:
    sns.regplot(x="Life expectancy at birth (years)", y="GDP", data=data[data['Country'] == country]);
    plt.title('Relationship between GDP and Life expectancy in {country}'.format(country=country))
    plt.xlabel('Life expectancy at birth (years)')
    plt.ylabel('GDP')
    plt.show()
    r, pval = stats.pearsonr(data['GDP'][data['Country'] == country], data['Life expectancy at birth (years)'][data['Country'] == country])
    print('The correlation coefficient bewtween GDP and Life expectancy in {country} is {r}'.format(country=country, r=r))
    
```

# Below is the average life expectancy in each country

```python
print('The average life expectancy in each country is: ')
data.groupby('Country')['Life expectancy at birth (years)'].mean()
```

```python
ax = sns.barplot(x='Country', y ='Life expectancy at birth (years)', data = data, errorbar= None)
plt.xticks(rotation=30, fontsize=8)
plt.title("Average Life Expectancy")
plt.show()
```

# What is the distribution of that life expectancy?

```python
for country in data_countries:
    plt.hist("Life expectancy at birth (years)",data = data[data['Country'] == country])
    plt.title(country)
    plt.xlabel('Year')
    plt.ylabel('GDP')
    plt.show()
```

# Size of each nations GDP

```python
sns.boxplot(x='Country', y='GDP', data=data)
plt.xticks(rotation=30)
plt.title('Difference in GDP')
plt.show()
```

# Conclusion
Although GDP size of each country in the data differs greatly with USA being the largest and Zimbabwe being the smallest, all the countries in the data experienced an increse in GDP between 2000 and 2015. 

China saw the most significant increase in GDP between 2000 and 2015 with GDP growth of 813.42 percent. Chile had a GDP growth of 211.48 percent. Zimbawbwe had a GDP growth of 143.72 percent. While USA, Germany and Mexico had a GDP growth of 75.73, 73.11, and 68.55 percent respectively.

In all the countries in the data, there is a very strong linear relationship between GDP and the life expectancy. Germany has the highest life expectancy with 79.7 and Zimbabwe has the lowest life expectancy with just 50.1.
